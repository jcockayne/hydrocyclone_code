```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.stats import gaussian_kde
import matplotlib
import matplotlib.lines as mlines
import eit
plt.rc('text', usetex=True)
plt.rc('font', family='serif', serif='cm')
```

```python
ns = [6,7,8,9,12]
n_bdy = 32
n_sensor = 8
```

```python
n_circ = 500
circle = eit.pol2cart(np.c_[np.ones(n_circ), np.linspace(-np.pi,np.pi,n_circ)])
```

```python
figures = [plt.figure(figsize=(4,4)) for _ in range(5)]
for n, fig in zip(ns, figures):
    a = fig.gca()
    grid = eit.grids.construct_circular(n, n_bdy, n_sensor)
    a.plot(circle[:,0], circle[:,1], c='gray', zorder=-100, alpha=0.5)
    
    a.set_aspect('equal')
    m_int = a.scatter(grid.interior_plus_boundary[:,0], grid.interior_plus_boundary[:,1], marker='x', label='$x_i^A$')
    m_bdy = a.scatter(grid.sensors[:,0], grid.sensors[:,1], marker='o', label='$x_i^B$')
    a.get_yaxis().set_ticks([])
    a.get_xaxis().set_ticks([])
a.legend([m_int, m_bdy], ['$x_i^A$', '$x_i^B$'], loc=1, fontsize=16)

for n, fig in zip(ns, figures):
    grid = eit.grids.construct_circular(n, n_bdy, n_sensor)
    actual_n = len(grid.all)
    fig.savefig('figures/design_n={}.pdf'.format(actual_n), bbox_inches='tight', transparent=True)
```
